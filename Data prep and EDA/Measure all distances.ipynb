# The distance will be used as a threshold for the recommendation

```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('whitegrid')
```

```python
users = pd.read_csv("userprofile.csv")
restaurants = pd.read_csv("geoplaces2.csv")
```

```python
users_cords=users[['userID', 'latitude', 'longitude']].copy()

```

```python
users_cords['latitudeU']=users_cords['latitude']
users_cords['longitudeU']=users_cords['longitude']
```

```python
users_cords.drop(['latitude', 'longitude'],inplace=True,axis=1)
```

```python
restaurants_cords=restaurants[['placeID', 'latitude', 'longitude']].copy()
```

```python
users_cords['key'] = 1
restaurants_cords['key'] = 1
```

```python
distances = pd.merge(users_cords,restaurants_cords,on='key').drop('key',axis=1)
```

```python
from geopy.distance import geodesic
```

```python
def distancer(row):
    coords_1 = (row['latitudeU'], row['longitudeU'])
    coords_2 = (row['latitude'], row['longitude'])
    return geodesic(coords_1, coords_2).km
```

```python
distances['distance']=distances.apply(distancer,axis=1)
```

```python
distances.head()
```

```python
distances['userID']=distances['userID'].apply(lambda x:x.split('U')[1])
```

```python
distances.to_csv(r'C:\Users\hedi\Desktop\Graduation project\AllPossibleDistances.csv', index = None, header=True)
```
